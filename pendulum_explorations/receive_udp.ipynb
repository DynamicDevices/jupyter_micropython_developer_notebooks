```python
import numpy
Gx, Gy, Gz = m.gx.mean()/100, m.gy.mean()/100, m.gz.mean()/100
gval = numpy.sqrt(Gx**2 + Gy**2 + Gz**2)
dp = (m.gx*Gx + m.gy*Gy + m.gz*Gz)/100/gval**2
m["theta"] = numpy.arccos(numpy.minimum(1, dp))
m.head()
```

```python
import sympy as sp
from sympy.interactive import printing
printing.init_printing(use_latex=True)
```

```python
g = sp.Symbol("g", positive=True)
t = sp.Symbol("t", real=True)
l = sp.Symbol("l", positive=True)
#theta0 = sp.Symbol("theta_0", positive=True)
theta = sp.Function("theta")(t)
theta
```

```python
q = sp.diff(sp.diff(theta)) + g/l * sp.sin(theta)
q
```

```python
q = sp.diff(sp.diff(theta)) + g/l * theta
H = sp.dsolve(q, theta)
H
```

```python
type(H)
```

```python
q = sp.Derivative(sp.Derivative(theta)) + g/l * sp.sin(theta)
q
H = sp.dsolve(q, theta)
H
```

```python
q = sp.Derivative(sp.Derivative(theta)) + g/l * theta
q
H = sp.dsolve(q, theta)
H
```

```python
H.rhs
```

```python
k = sp.diff(sp.diff(H.rhs, t), t) + g/l*H.rhs
k.simplify()
```
