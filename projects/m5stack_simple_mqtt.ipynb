```micropython
#%sendtofile main.py

import machine, time, network

fconfig = dict(x.split()  for x in open("config.txt"))
wifiname = fconfig["wifiname"]
wifipassword = fconfig["wifipassword"]
mqttbroker = fconfig["mqttbroker"]
boardname = fconfig["boardname"]

w = network.WLAN(network.STA_IF)
w.active(True)
w.connect(wifiname, wifipassword)
while not w.isconnected():
    time.sleep(0.1)
print("connected to wifi")

from umqtt.robust import MQTTClient
client = MQTTClient(boardname, mqttbroker, 1883)
for i in range(100):
    try:
        client.connect()
        break
    except OSError as e:
        print(e)
print("connected to mqtt")

client.publish(boardname+"/ip", w.ifconfig()[0])

# To listen on the command line of your PC, do:
# mosquitto_sub -h mqtt.local -v -t "m5stack1/#"

from m5stack import lcd

topiccolours = b"esp32colours/crgb"

def ijgen():
    while True:
        for j in range(5, 240, 10):
            for i in range(5, 320, 10):
                yield i,j
                
ijg = ijgen()
maxv = 1

def callback(topic, msg):
    global maxv
    print((topic, msg))
    i, j = next(ijg)
    cc = list(map(int, msg.split()[1:]))
    maxv = max(maxv, max(cc))
    col = sum(int(a*255/maxv)<<b  for a, b in zip(cc, [16, 8, 0]))
    lcd.circle(i, j, 8, 0, col)
        
client.set_callback(callback)
client.subscribe(topiccolours)

while True:
    client.wait_msg()
```
