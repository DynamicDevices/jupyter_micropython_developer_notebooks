```python
from matplotlib import pyplot as plt
%matplotlib inline

def plotdt(fname, title):
    k = open(fname).read()
    k = [ int(k[i-8:i])  for i in range(8, len(k), 8) ]
    k = [a-b  for a,b in zip(k[1:], k)]
    plt.plot(k[2:])
    plt.ylabel("ms")
    plt.title(title)
    return k
k = plotdt("data15.txt", "Immediate save to SD")
```
